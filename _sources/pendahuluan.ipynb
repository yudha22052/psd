### Latar Belakang

Persaingan dalam dunia bisnis dan tingginya permintaan pasar, khususnya di sektor pertanian terkait produk pupuk, mendorong perusahaan untuk menjaga stabilitas dalam berbagai aspek produksi. Namun, sering kali ditemukan kendala yang tidak sesuai dengan kebutuhan, terutama terkait stok bahan baku untuk produksi pupuk. Masalah ini muncul karena perusahaan belum menerapkan sistem khusus untuk meramalkan kebutuhan stok bahan baku di masa depan. Saat ini, perusahaan hanya menyesuaikan pembelian stok bahan baku berdasarkan jumlah produksi pupuk di periode tertentu, yang menyebabkan terjadinya kelebihan atau kekurangan stok di waktu tertentu. Padahal, perusahaan memiliki data yang dapat diolah lebih lanjut untuk menghasilkan keuntungan, salah satunya dengan memprediksi kebutuhan stok bahan baku di masa mendatang melalui proses peramalan.

Peramalan (forecasting) merupakan proses memprediksi nilai suatu variabel berdasarkan data historis atau variabel terkait. Ini adalah pendekatan kuantitatif yang digunakan untuk memperkirakan kejadian di masa depan dengan menggunakan data masa lalu sebagai referensi. Salah satu manfaat utama peramalan adalah memprediksi penjualan secara akurat dari waktu ke waktu, sehingga jumlah stok bahan baku yang diperlukan oleh perusahaan dapat diantisipasi dengan lebih baik.

Berdasarkan masalah ini, diperlukan penerapan peramalan stok bahan baku pupuk dengan menggunakan metode yang tepat. Beberapa metode yang umum digunakan dalam peramalan antara lain moving average, exponential smoothing, ARIMA, dan NaÃ¯ve. Penelitian ini akan mengimplementasikan metode Triple Exponential Smoothing dalam Business Intelligence untuk memprediksi kebutuhan stok bahan baku pupuk di periode mendatang. Diharapkan penelitian ini dapat membantu perusahaan dalam proses peramalan penjualan guna menentukan kebutuhan stok bahan baku secara lebih akurat.

### Tujuan

Proyek ini bertujuan untuk mendukung tercapainya visi PT Petrokimia sebagai pemain dominan dalam solusi agro dan bahan kimia industri yang terintegrasi melalui penerapan analisis data yang efektif. Analisis data akan digunakan untuk:
1. Melalui peramalan stok bahan baku menggunakan metode Triple Exponential Smoothing, perusahaan dapat mengoptimalkan manajemen stok bahan baku pupuk. Hal ini akan membantu perusahaan mendukung penyediaan pupuk nasional, meningkatkan ketahanan pangan, dan mencapai visi misi perusahaan yang sejalan dengan Sustainable Development Goals (SDGs).
2.	Analisis data memungkinkan perusahaan untuk merancang strategi yang lebih responsif terhadap perubahan permintaan pasar dan fluktuasi produksi. Dengan prediksi stok yang lebih akurat, perusahaan dapat meningkatkan efisiensi operasional dan menyesuaikan proses produksi dengan kebutuhan pasar secara tepat waktu.
3.	Membantu perusahaan menghindari kelebihan dan kekurangan stok bahan baku, sehingga mengurangi biaya penyimpanan dan risiko produksi terhenti. Ini akan membantu perusahaan mencapai target bisnis berupa peningkatan produktivitas, pengurangan biaya operasional, dan peningkatan kontribusi terhadap kemajuan industri kimia nasional.
4.	Dengan menggunakan analisis data untuk perencanaan produksi dan manajemen stok, perusahaan dapat memanfaatkan sumber daya manusia yang tangguh dan inovatif untuk mengembangkan proses bisnis yang lebih efektif dan efisien, sesuai dengan misi perusahaan.
